# Image Thresholding 

'Process for separating the image from its background. In this threshold value is provided and that value is compared with each pixel value of the image and results the output'

```python
img = cv.imread("Images/crick.jpg",0)
_,th1 = cv.threshold(img,127,220,cv.THRESH_BINARY)      
# Pixels values less than 127 become 0 and greater than 127 becomes 1 but less than 220
_,th2 =  cv.threshold(img,127,220,cv.THRESH_BINARY_INV)    # inverse the result of above thresh_binary
_,th3 = cv.threshold(img,100,200,cv.THRESH_TRUNC)  # values remain same upto 100 and after that it becomes 100 for all the pixels lies in range.
_,th4 = cv.threshold(img,100,200,cv.THRESH_TOZERO) # values becomes 0 for less than 100 and after that it remains same for a given range.
_,th5 = cv.threshold(img,100,200,cv.THRESH_TOZERO_INV)  # inverse of TOZERO
_,th6 = cv.threshold(img,100,200,cv.THRESH_TRIANGLE)

# cv.imshow("Image",img)
# cv.imshow("Threshold1",th1)
# cv.imshow("Threshold2",th2) 
# cv.imshow("Threshold3",th3) 
# cv.imshow("Threshold4",th4)
# cv.imshow("Threshold5",th5)
# cv.imshow("Threshold6",th6)

titles =  ["THRESH_BINARY","THRESH_BINARY_INV","THRESH_TRUNC","THRESH_TOZERO","THRESH_TOZERO_INV","THRESH_TRIANGLE"]
images =   [th1,th2,th3,th4,th5,th6]

plt.figure(figsize=(14,5))
for x in range(6):
    plt.subplot(2,3,x+1)
    plt.imshow(images[x],cmap="viridis")
    plt.title(titles[x])
    plt.autoscale(axis='y')


if( cv.waitKey(0) & 0xFF == ord('q') ):
    cv.destroyAllWindows()
```

# Morphological Transformations

-> Performed on binary images
-> 2 things are Required: original Image and a Kernel
-> Kernel tells how to change the value of a given pixel by combining it with different amount of Neighbouring Pixels

```python
img = cv.imread("Images/candies.jpg",0)
_,th1 = cv.threshold(img,180,255,cv.THRESH_BINARY)

kernel = np.ones((5,5),np.uint8)

dilation = cv.dilate(th1,kernel,iterations=20)    # opposite of erosion
erosion = cv.erode(th1,kernel,iterations=20)     # Erodes the boundary of the foreground object or removes the small white noises
opening = cv.morphologyEx(th1,cv.MORPH_OPEN,kernel)  # Erosion first and then dilation
closing = cv.morphologyEx(th1,cv.MORPH_CLOSE,kernel)   # Dilation first and then erosion
gradient = cv.morphologyEx(th1,cv.MORPH_GRADIENT,kernel)  # It is the difference between dilation and erosion of an image.
tophat = cv.morphologyEx(img, cv.MORPH_TOPHAT, kernel)  # difference between input image and Opening of the image.


images = [img,th1,dilation,erosion,opening,closing,gradient,tophat]
names = ["Image", "Threshold Binary Inverse", "dilation","erosion","opening","closing","gradient","tophat"]

plt.figure(figsize=(10,8))
for x in range(8):
    plt.subplot(4,2,x+1)
    plt.imshow(images[x],"gray")
    plt.title(names[x])
```

# Motion Detection and Tracking 

```python
cap = cv.VideoCapture("pedestrianWalk.mp4")
_,frame10 = cap.read()
_,frame11 = cap.read()

#print(frame1.shape , frame2.shape,type(frame1))

# cv.imshow("hj",frame1)
# cv.imshow("h1",frame2)
# cv.waitKey(0)
count=0

while(cap.isOpened()):
    
    differnce = cv.absdiff( frame10, frame11)        # Absolute differnce between 2 Frames
    grayScale = cv.cvtColor(differnce,cv.COLOR_BGR2GRAY)   # Convert the Frames into Gray Scale
    gaussian  = cv.GaussianBlur(grayScale,(5,5),0)                 # Smoothening Of image
    _,th1 = cv.threshold(gaussian,40,255,cv.THRESH_BINARY)      
    dilation = cv.dilate(th1,None,iterations=3)
    
    contours,heirachy = cv.findContours(dilation,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
    
    for i in contours:
        x,y,width,height = cv.boundingRect(i)
        
        contour_area = cv.contourArea(i)
        
        if(contour_area < 700):
            continue
        count+=1
        cv.rectangle(frame10,(x,y),(x+width,y+height),(0,255,0),2)
        cv.putText(frame10,"Status: {} ".format("Movement"),(10,10),cv.FONT_ITALIC,1.5,(255,0,0),2,cv.LINE_AA)
        cv.putText(frame10,"Count: {} ".format(str(count)),(20,10),cv.FONT_ITALIC,1.5,(0,255,0),2,cv.LINE_AA)
        
#     cv.drawContours(frame1,contours,-1,(0,0,255),2)
    
    
#     cv.imshow("Detection",frame1)
    frame10 = frame11
    _,frame11 = cap.read()
    
cap.release()
cv.destroyAllWindows()
    
    
```

# Detect Simple Geometric Shapes

```python
img = cv.imread("Images/shape.png")
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
_,th1 = cv.threshold(gray,200,255,cv.THRESH_BINARY)
contours,_= cv.findContours(th1,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)

for x in contours:
    approx = cv.approxPolyDP(x,0.01 * cv.arcLength(x,True),True )
    print(approx,"and")
    # approxPolyDP:- Approximates a polygonal curve(s) with the specified precision.
    # arcLength:- Calculates a contour perimeter or a curve length. See Docs
    cv.drawContours(img,[approx],0,(0,255,0),2)
    # [approx ] tells the contor array we get of x,y coordinates
    
    x = approx.ravel()[0]                  # Ravel is responible for flatten the array
    y = approx.ravel()[1]-1
    
    print(x,y)
    
    if( len(approx)==3):
        cv.putText(img,"Triangle",(x,y),cv.FONT_ITALIC,1,(0,0,255),1,cv.LINE_AA)
    
    elif( len(approx)==4):
        x1,y1,w,h = cv.boundingRect(approx)
        aspectRatio = float(w)/h
        
        if(aspectRatio >=0.95 and aspectRatio<=1.045):
            cv.putText(img,"Square",(x,y),cv.FONT_ITALIC,1,(0,0,255),1,cv.LINE_AA)
        
        else:
            cv.putText(img,"Rectangle",(x,y),cv.FONT_ITALIC,1,(0,0,255),1,cv.LINE_AA)
            
    elif( len(approx)==5):
        cv.putText(img,"Pentagon",(x,y),cv.FONT_ITALIC,1,(0,0,255),1,cv.LINE_AA)
    
    elif( len(approx)==6):
        cv.putText(img,"Star",(x,y),cv.FONT_ITALIC,1,(0,0,255),1,cv.LINE_AA)
    
    else:
        cv.putText(img,"Polynomial",(x,y),cv.FONT_ITALIC,1,(0,0,255),1,cv.LINE_AA)
    
    
cv.imshow("image",img)
cv.imshow("th1",th1)
cv.waitKey(0)
```

# Hough Transform Basics
1--> The Hough Line Transform is a transform used to detect straight lines.
2--> It is of 2 types:  Standard Hough Transform(Hough Lines Method) and Probabilistic Hough Line Transform( HoghLines P Method)

```python
img = cv.imread("Images/Lanes.jpg")
gray  =cv.cvtColor(img,cv.COLOR_BGR2GRAY)
canny = cv.Canny(gray,50,50)
line1 = cv.HoughLines(canny,1,np.pi/180,200)                 # Disadvantage - Forms the Infinite Lines
line2 = cv.HoughLinesP(canny,2,np.pi/180,300,100,10)


# ----------------------------------------------- Hough Lines ----------------------------------------
# for pt in line:
#     rho,theta = pt[0]
#     a = np.cos(theta)
#     b  =np.sin(theta)
#     x0 = a * rho
#     y0 = b * rho
    
#     x1 = int(x0 - 1000 * (b))
#     y1 = int(x0 + 1000 * (a))
    
#     x2 = int( x0 + 1000 * (b) )
#     y2  =int( x0 - 1000 * (a))
# --------------------------------------------------------------------------------------------------------------------------

#------------------------------------------------------- Hough Line P ------------------------------------------------------
for line in line2:
    x1,y1,x2,y2 = line[0]
    cv.line(img,(x1,y1),(x2,y2),(0,255,0),3)    
#--------------------------------------------------------------------------------------------------------------------------    


cv.imshow("Image",img)
cv.imshow("Canny",canny)
if( cv.waitKey(0) & 0xFF == ord('q') ):
    cv.destroyAllWindows()
```

# Lane Detection in an Image

```python
img = cv.imread("Images/Lanes.jpg")
gray = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.figure(figsize=(15,5))
height,width,_=img.shape
# plt.imshow(gray)
# region_of_area = [(0,height),(100,210),(400,height) ]
region_of_area = [(0,height),(0,250),(185,100),(265,100),(360,300),(400,450)]
#---------------------------------- Region of Interest ------------------------------------------------------

def region_of_interest(newimg,newVertice):
    
    mask = np.zeros_like(newimg)
    masked_channel = 255                              
    cv.fillPoly(mask,[newVertice],masked_channel)
    masked_image = cv.bitwise_and(mask,newimg)
#     cv.imshow("IMg1",mask)
#     cv.imshow("IM2",masked_image)
#     cv.waitKey(0)
    return masked_image
#---------------------------------------------------------------------------------------------------------------

def draw_lines(img,lines):
    img1 = np.copy(img)
    blank_img = np.zeros((img.shape),dtype=np.uint8)
    
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            
            cv.line(img1,(x1,y1),(x2,y2),(0,255,0),2)
            
        
    return img1


gray_convert = cv.cvtColor(img,cv.COLOR_RGB2GRAY)
edge_detection = cv.Canny(gray_convert,80,150)
cropped_image = region_of_interest(edge_detection,np.array(region_of_area,np.int64))


line2 = cv.HoughLinesP(cropped_image,5,np.pi/60,100,lines=np.array([]),minLineLength=20,maxLineGap=5)

lane_detect = draw_lines(img,line2)

     

# cv.imshow("Cropped",edge_detection)
cv.imshow("Image",lane_detect)



    
    

if( cv.waitKey(0) & 0xFF == ord('q') ):
    cv.destroyAllWindows()
```

# Circle Detection using the Image

```python
img = cv.imread("Images/football.jpg")
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
canny = cv.Canny(gray,20,100)
blur = cv.medianBlur(gray,5)

circles = cv.HoughCircles(blur,cv.HOUGH_GRADIENT,dp=1,minDist=600,param1=40,param2=20,minRadius=200,maxRadius=400)
# print(circles,type(circles))                   # Return x,y,r in array form.

for coord in circles:
    for x,y,r in coord:
        cv.circle(img,(x,y),r,(0,255,255),2,cv.LINE_AA)
        cv.circle(img,(x,y),4,(255,255,0),-1,cv.LINE_8)


cv.imshow("Real",img)
# cv.imshow("Images",blur)
# cv.imshow("Canny",canny)
if( cv.waitKey(0) & 0xFF == ord('q') ):
    cv.destroyAllWindows()
```

# Shi-Tomashi Corner Detection Method

```python
img = cv.imread("Images/img2.jpg")
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
coord = cv.goodFeaturesToTrack(gray,100,0.1,5)
# print(coord,len(coord))

for pts in coord:
#     print("and",pts.ravel(),pts[1])
    x,y = pts[0]
    cv.circle(img,(x,y),3,(0,0,255),-1)
    
cv.imshow("Image",img)    

if( cv.waitKey(0) & 0xFF == ord('q') ):
    cv.destroyAllWindows()
```
